```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python


# Veriyi yÃ¼kle
df = pd.read_csv("winequality-red.csv")

# Ä°lk 5 satÄ±r
print(df.head())

# Eksik veri kontrolÃ¼
print(df.isnull().sum())

# Ãzellikler arasÄ± korelasyon
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Korelasyon Matrisi")
plt.show()
```

```python
# Hedef deÄiÅkeni Ã§Ä±kar (clustering iÃ§in kullanmayacaÄÄ±z)
X = df.drop("quality", axis=1)

# Veriyi standardize et
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker="o")
plt.xlabel("KÃ¼me SayÄ±sÄ±")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()

# Silhouette Skoru
silhouette_scores = []
for k in range(2, 8):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)
    print(f"K={k}, Silhouette Skoru: {score:.3f}")
```

```python
# K=3 seÃ§elim (silhouette skoru ve yorumlanabilirlik iÃ§in)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# KÃ¼me sonuÃ§larÄ±nÄ± veriye ekle
df["cluster"] = clusters

# KÃ¼me daÄÄ±lÄ±mÄ±nÄ± gÃ¶rselleÅtir
sns.countplot(x="cluster", data=df)
plt.title("KÃ¼me DaÄÄ±lÄ±mÄ±")
plt.show()
```

```python
# KÃ¼me ortalamalarÄ±nÄ± incele
cluster_means = df.groupby("cluster").mean()
print(cluster_means)

# AlkollÃ¼ kÃ¼meleri karÅÄ±laÅtÄ±r
sns.boxplot(x="cluster", y="alcohol", data=df)
plt.title("Alkol Seviyesine GÃ¶re KÃ¼meler")
plt.show()
```

Cluster 0: DÃ¼ÅÃ¼k alkol, yÃ¼ksek asitlik.

Cluster 1: YÃ¼ksek alkol, dÃ¼ÅÃ¼k Åeker.

Cluster 2: Orta seviye Ã¶zellikler.

```python
from sklearn.decomposition import PCA

# PCA ile 2 boyuta indirge
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# PCA sonuÃ§larÄ±nÄ± dataframe'e ekle
df["PCA1"] = X_pca[:, 0]
df["PCA2"] = X_pca[:, 1]

# KÃ¼meleme sonuÃ§larÄ±nÄ± gÃ¶rselleÅtir
plt.figure(figsize=(10, 6))
sns.scatterplot(x="PCA1", y="PCA2", hue="cluster", data=df, palette="viridis")
plt.title("PCA ile KÃ¼meleme SonuÃ§larÄ±")
plt.show()
```
